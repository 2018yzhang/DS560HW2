# Run Three Scripts in One Notebook

This notebook executes three scripts, **GenerateNums.py**, **ExecuteFunction.py** and **Graph.py**, in order. It will also show the visualizations of the **initial**, **intermediate**, and **final results**. It shows the distribution of  the initial and intermediate results.

```python
import os
import csv
import seaborn as sns
import pandas as pd
from IPython.display import Image
```

### Execute GenerateNums.py and Visualize
The script generates 1000 numbers randomly and saves them in RandomNums.csv. Then draw its distribution using seaborn package.

```python
os.system('python GenerateNums.py')
```

```python
x = []
with open('RandomNums.csv','r') as xFile:
    xHeader = xFile.readline()
    xList = csv.reader(xFile)
    for row in xList:
        x.append(int(row[0]))

randomNumsDF = pd.DataFrame(x,columns =['RandomNums'])
sns.distplot(randomNumsDF['RandomNums'])
```
